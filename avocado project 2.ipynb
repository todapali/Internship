# Modifying the data in order to make useful predictions

```python
target_df = df['AveragePrice']
```

```python
df = df.drop('AveragePrice', axis = 1)
```

```python
target_df.head()
```

```python
df.head()
```

# Encoding the Categorical Variables

Encoding with Frequency Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit([df['region']])
df['region']= encoder.fit_transform(df[['region']])
```

```python
df.head()
```

# Encoding with 0 or 1

```python
df['type'] = pd.get_dummies(df['type'], drop_first = True)
```

```python
df.head()
```

```python
X = df.values
y = target_df.values
```

# Splitting the Dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

# Training the Model

```python
import xgboost as xgb
regressor = xgb.XGBRegressor(n_estimators = 800, max_depth = 10)
regressor.fit(X_train, y_train)
```

# Making Predictions

```python
y_pred = regressor.predict(X_test)
```

# Evaluating the Trained Model on the Test Set

R2 Score Evaluation (The more it is closer to 1, the better is the model)

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

Scatter Plot between the true labels and predicted labels

```python
fig10 = plt.figure(figsize = (10, 5))
axes6 = fig10.add_axes([0.1, 0.1, 0.8, 0.8])
sns.scatterplot(y_test, y_pred, linewidth = 1, ax = axes6)
plt.title('Scatter Plot between the true labels and predicted labels')
axes6.set_xlabel("True Average Price")
axes6.set_ylabel("Predicted Average Price")
```
